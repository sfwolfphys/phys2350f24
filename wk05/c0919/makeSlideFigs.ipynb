```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0,3,100)
vReal = np.tanh(t)
vIdeal = t

with plt.xkcd():
    plt.axhline(y=1,color='grey',linestyle='dashed')
    plt.plot(t,vIdeal,'r', label='Ideal case')
    plt.plot(t,vReal,'b', label='With air resistance')
    plt.legend()
    plt.xticks([])
    plt.yticks([])
    plt.xlabel('time')
    plt.ylabel('velocity')
    plt.savefig('velTimePlot.png')
    plt.show()
```
